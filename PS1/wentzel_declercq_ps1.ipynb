### 1.4 Simple statistical tests

Using a t-test, compare the median house value in locations that are adjacent to the Charles River to locations that are not adjacent to the Charles River. Do you reject the null hypothesis that prices are equivalent?

````python
# enter your code here
NextToScharles = bdataf.loc[ (bdataf["CHAS"] == 1)]["MedianPrice"]
NotNextToScharles = bdataf.loc[ (bdataf["CHAS"] == 0)]["MedianPrice"]

print(NextToScharles.shape)
print(NotNextToScharles.shape)
#use scipy.ttest
#We have different sample sizes so we have to set equal_var = False and this uses Welsch's t-test
#Welsch's t-test says p < 0.05
print(sp.stats.ttest_ind(NextToScharles.values,NotNextToScharles.values,equal_var=False))
````

*Write your response here.*
The p-value is 0.00356 for a t-test performed on two samples with unequal size (hence equal_var = False)). Since this p-value is less than 0.05, which is considered to be the statistically significant value, we will reject the null hypothesis that the average of the median prices of neighborhood next to the river is the same as the average of the median prices of neighborhoods not next to the river. Thus we accept the alternative hypothesis which is that the averages of the media prices are different.
